```python
import requests
#fetches the link we will need
wiki_link = "https://en.wikipedia.org/wiki/List_of_Asian_countries_by_population"
#url of list of asian countries on wikipedia
link = requests.get(wiki_link).text
#converts the link to text when returned to us
```

```python
#print(link)
```

```python
from bs4 import BeautifulSoup as BS
soup =BS(link,'lxml')
#beautiful soup lets us work with the html from the wiki page
```

```python
#table class is  = 'sortable wikitable sticky-header col2left' when you inspect it on the wikipedia. It can be different over time
```

```python
target_table = soup.find('table',class_= 'sortable wikitable sticky-header col2left')
#print(target_table)
```

```python
table_links = target_table.findAll('a')
#table_links
```

```python
countries = [] # list that holds the countries in the table
for links in table_links: #for all items in the table links we found
    countries.append(links.get('title')) #add the title of those links i.e country that's hyperlinked to a reference, to our empty list called "Countries'
print(countries)
```

```python
#convert our country list to a dataframe so we can do some data manipulation using pandas
import pandas as pd
df = pd.DataFrame()
df['Countries'] = countries
df
```

```python
df = df.dropna() #drop the none/null values we ar seeing
df
```

```python
countries = df["Countries"]
df[countries.isin(countries[countries.duplicated()])].sort_values("Countries")
#China is duplicated two times after the first as it has several territories but the scraper catches it as another country
#for simplicity, we will prune all 'china' values except the first one
```

```python
df = df.drop_duplicates(keep ='first')
```
